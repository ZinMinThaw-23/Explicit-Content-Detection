Data Evaluation

```python
test_data=pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
test_data
```

```python
test_data.describe()
```

```python
for i in test_data.columns:
  print(i)
  print(test_data[i].isnull().unique())
  print()
```

```python
X_test=test_data.iloc[:,:-1]
Y_test=test_data.iloc[:,-1]
```

```python
Y_pred1=model1.predict(X_test)
Y_pred2=model2.predict(X_test)
```

```python
def meansquareerror(Y_pred,Y_test):
  for i in range(len(Y_pred)):
    total=0
    SE=(Y_pred-Y_test)**2
    total+=SE
  MSE=total/len(Y_pred)
  print("Mean Square Error for this regression is",MSE)
  return MSE
```

```python
LG_MSE=meansquareerror(Y_pred1,Y_test)
```

```python
DT_MSE=meansquareerror(Y_pred2,Y_test)
```

```python
MSE_difference=LG_MSE-DT_MSE
print("Mean Square Error Difference is",MSE_difference)
```
